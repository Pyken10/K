# FOOTBALL PLAYER MANAGEMENT SYSTEM

```python
def add():
    pid = input("Player ID: ")
    pname = input("Player Name: ")
    ppos = input("Player Position: ")
    pteam = input("Player Team: ")
    with open("players.txt", "a") as f:
        f.write(pid + "\t")
        f.write(pname + "\t")
        f.write(ppos + "\t")
        f.write(pteam + "\n")
        print("Player information added.")

def show():
    with open("players.txt", "r") as f:
        print("ID\t Name\t Position\t Team")
        print(f.read())

def search():
    i = input("Enter the ID of the player to search: ")
    with open("players.txt", "r") as f:
        all = f.readlines()
        for data in all:
            d = data.split("\t", 1)
            if d[0] == i:
                print(data)

def delete():
    i = input("Enter the ID of the player to delete: ")
    with open("players.txt", "r") as f:
        all = f.readlines()
    with open("players.txt", "w") as f:
        for data in all:
            d = data.split("\t", 1)
            if d[0] != i:
                f.writelines(data)
        print("Player information deleted.")

def update():
    i = input("Enter the ID of the player to update: ")
    with open("players.txt", "r") as f:
        all = f.readlines()
    with open("players.txt", "w") as f:
        for data in all:
            d = data.split("\t", 1)
            if d[0] == i:
                nn = input("Enter new name: ")
                np = input("Enter new position: ")
                nt = input("Enter new team: ")
                f.writelines(d[0] + "\t" + nn + "\t" + np + "\t" + nt + "\n")
            else:
                f.writelines(data)
        print("Player information updated.")

while True:
    print("Welcome to the Football Player Management System.")
    print("1. Add player information")
    print("2. Delete player information")
    print("3. Update player information")
    print("4. Search player information")
    print("5. Show all player information")
    print("6. Exit")
    ch = int(input("Enter your choice: "))
    if ch == 1:
        add()
    if ch == 2:
        delete()
    if ch == 3:
        update()
    if ch == 4:
        search()
    if ch == 5:
        show()
    if ch == 6:
        break;
```
